```python
for file_index in range(6):
    start = time.time()
    file = open("testcases/test"+str(file_index)+".txt", "r")
    lines = file.read().split('\n')
    M=int(lines[0].split(' ')[0])
    T=int(lines[0].split(' ')[1])
    timerequired=str(max(20, 0.1*(M + 1))/1000)
    model=[]
    for i in lines[1:-1]:
        model.append([int(q) for q in i.split(' ')[1:]])
    model=[list(set(t)) for t in model]
    oldmodel=model.copy()
    print("===============================================")
#     print("M: "+str(M))
#     print("T: "+str(T))
#     print("Model: "+str(model))
#     print("Worse case total money: "+str(sum([min(sl) for sl in model])))
    
    
    # Greedy Algorithm
    flat_list = [item for sublist in model for item in sublist]
    # print(flat_list)
    # Sort O(n log n)
    flat_list.sort()
    out=[]
    loc=[]
    
    while len(model)>0:

        while True:
            current_selection=flat_list[-1]
            s=0
    # Find which type to remove
            dellist=0
            nextmin=float('-inf')
            for sublist in range(len(model)):
                if current_selection in model[sublist]:
                    rest=[i for i in model[sublist] if i!= current_selection]
                    if len(rest)!=0:
                        tmpmax=min(rest)
                        if tmpmax>nextmin:
                            nextmin=tmpmax
                            dellist=sublist
                    else:
                        tmpmax=current_selection
                        if tmpmax>nextmin:
                            nextmin=tmpmax
                            dellist=sublist

    # Find which type to remove end

            for sl in range(len(model)):
                if dellist!= sl:
                    s=s+min(model[sl])
            if current_selection+s>M:
                for sl in model:
                    while current_selection in sl:
                        sl.remove(current_selection)
                        flat_list.remove(current_selection)
            else:
                break
        loc.append(dellist)
        for i in model[dellist]:
            flat_list.remove(i)
        model.remove(model[dellist])
        out.append(current_selection)
        M=M-current_selection

    a=[i for i in range(T)]
    locs=[]
    for i in loc:
        locs.append(a[i])
        a.remove(a[i])
    output=[]
    for i in range(T):
        output.append(oldmodel[i].index(out[locs.index(i)]))

#     print("Money output: "+str(out))
    print("Money: "+str(sum(out)))
    print("output:"+str(output))
    
    end = time.time()
    print("Time: "+ str(end - start))
    print("Time required: "+timerequired)
    print("Time passed: "+str((end - start)<float(timerequired)))
    file = open("testcases/out"+str(file_index)+".txt", "r")
    lines = file.read().split('\n')
    print("Actual Money output:"+str(lines[0]).split(' ')[0])
    print("Actual Required output:"+str(str(lines[0]).split(' ')[1:]))
```
